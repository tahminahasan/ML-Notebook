```python
df=pd.read_csv("/content/appartment_price.csv")
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df.drop('Price', axis =1), df.Price)
```

```python
model.predict([[1600, 3 , 2]])
```
