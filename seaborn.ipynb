```python
import seaborn as sns
import pandas as pd
from IPython.display import Image
```

```python
df = sns.load_dataset('penguins')
#df = sns.load_dataset('tips')# another dataset
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
sns.histplot(data=df,x='bill_depth_mm')
```

```python
sns.histplot(data=df,x='bill_depth_mm', hue='island')
```

```python
sns.histplot(data=df,x='bill_depth_mm', kde= True)
```

```python
sns.histplot(data=df, x='bill_depth_mm', kde= True, hue='sex')
```

```python
#joint plot
sns.jointplot(data=df,x='bill_length_mm', y='flipper_length_mm')
```

```python
sns.jointplot(data=df,x='bill_length_mm', y='flipper_length_mm', kind= 'reg')
```

```python
sns.jointplot(data=df,x='bill_length_mm', y='flipper_length_mm', kind= 'scatter')
```

```python
sns.jointplot(data=df,x='bill_length_mm', y='flipper_length_mm', kind= 'hex')
```

```python
#pair plot
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,kind ='reg')
```

```python
#categorical data
#countplot
sns.countplot(data=df, x= 'island', palette='PuBu')
```

```python
#need both categorical and numerical
sns.barplot(df, y= 'flipper_length_mm', x='sex', hue='island')
```

```python
#boxplot
sns.boxplot(df, y= 'flipper_length_mm', x='sex', hue='island')
```

```python
sns.violinplot(df, y= 'flipper_length_mm', x='sex')
```

```python
sns.violinplot(data=df, x="species", y="flipper_length_mm", fill=False)
```

```python
#pair grid
g=sns.PairGrid(df)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(df)
g.map(sns.scatterplot)
g.map_diag(sns.histplot)
```
