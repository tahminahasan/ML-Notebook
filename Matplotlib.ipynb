```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x= [10,20,30,40,50,60]
y=[100,200,300,400,500,300]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color= 'red', linestyle= '--', linewidth=10)
```

```python
plt.plot(x,y, 'b-.o')
```

```python
plt.xlabel("Price")
plt.ylabel("amount")
plt.title("Matlab chart")
plt.plot(x,y, 'b-.o')
```

```python
#multiple data
x1= [10,20,30,40,30]
x2= [15,25,30,35,40]
y1=[100,200,300,350,400]
y2=[25,35,45,55,65]
y3= [100,150,200,250,300]
```

```python
plt.xlabel("Price")
plt.ylabel("amount")
plt.title("Matlab chart")
plt.plot(x1,y1,'*r--', label='Apple')
plt.plot(x1,y2,color='y', label='Banana')
plt.plot(x2,y3,color='orange', label='Mango')
plt.legend(loc='best', shadow=True, fontsize='large')
plt.grid()
```

```python
#subplot
x1= [10,20,30,40,30]
x2= [15,25,30,35,40]
y1=[100,200,300,350,400]
y2=[25,35,45,55,65]
y3= [100,150,200,250,300]
plt.xlabel("Price")
plt.ylabel("amount")
plt.title("Matlab chart")

#(2,2,1) here 2,2 means 2 by 2 and 1 is the position
plt.subplot(2,2,1)
plt.plot(x1,y1,'*r--', label='Apple')
plt.grid()
plt.legend(loc='best', shadow=True, fontsize='large')

plt.subplot(2,2,2)
plt.plot(x1,y2,color='y', label='Banana')

plt.subplot(2,2,4)
plt.plot(x2,y3,color='orange', label='Mango')
plt.grid()
```

```python
#bar chart
name = ['A', 'B', 'C', 'D', 'E']
id = [201,202,203,204,205]
marks = [90,95,83,75,86]
```

```python
plt.bar(name,id)
```

```python
plt.bar(name,id)
plt.bar(name,marks)
```

```python
plt.bar(name,id, label='roll', width=0.3)
plt.bar(name,marks, label='marks', width=0.3)
```

```python
xpos= np.arange(len(name))
plt.bar(xpos-0.2,id, label='id', width=0.3, color = 'black')
plt.bar(xpos+0.2,marks, label='marks', width=0.3, color= 'blue')
plt.legend()
```

```python
#histogram
number = [45,55,65,45,55,75,70,100,150,120,130,120,130]
plt.hist(number)
```

```python
plt.hist(number, bins =[45,55,65,75, 85,90,100, 120, 130, 140 ,150,200], rwidth =0.8, histtype ='bar', color= 'black')
```

```python
men= [100,200,150,300,250,270,400]
women=[200,230,240,270,400,340]
plt.hist([men,women], bins=[50,60,100,120,200,250,270,300,350,400], color = ['blue', 'pink'], label = ['men', 'women'], rwidth = 0.50)
plt.legend()
```

```python
#pie chart
expense =[ 2500,1000,700,500]
exp_name= ['food','rent', 'bill', 'others']
colors = ['r', 'g', 'b' , 'y']
plt.pie(expense, labels =exp_name)
```

```python
expense = [ 2500,1000,700,500]
exp_name= ['food','rent', 'bill', 'others']
colors = ['r', 'g', 'b' , 'y']
explode =[0.0,0.0,0.0,0.2]
plt.pie(expense, labels =exp_name, colors=colors, explode=explode, autopct = "%.2f%%")
plt.legend()
```
